```python
# NOTE: 2017 is NOT a leap year
def day_pass(date_month, date_day):
    month = int(date_month)
    day = int(date_day)
    
    if month in [1, 3, 5, 7, 8, 10]: # Big month
        if day < 31:
            day += 1
        else:
            day = 1
            month += 1
    elif month == 2: # February
        if day < 28:
            day += 1
        else:
            day = 1
            month += 1
    elif month == 12: # December, 12/31 -> 1/1
        if day < 31:
            day += 1
        else:
            day = 1
            month = 1
    else: # Small month
        if day < 30:
            day += 1
        else:
            day = 1
            month += 1
    
    if day < 10:
        day = '0' + str(day)
    else:
        day = str(day)
    
    if month < 10:
        month = '0' + str(month)
    else:
        month = str(month)
        
    return [month, day]
    
```

```python
def clean(data_json_parsed):
    l = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
    if data_json_parsed['feeds'] == []:
        return []
    
    hour = int(list(data_json_parsed['feeds'][0]['AirBox'][0].keys())[0][11:13])
    l[hour] = data_json_parsed['feeds'][0]['AirBox'][0][list(data_json_parsed['feeds'][0]['AirBox'][0].keys())[0]]['s_d0']
    
    for data in data_json_parsed['feeds'][0]['AirBox']:
        new_hour = int(list(data.keys())[0][11:13])
        if new_hour > hour:
            hour = new_hour
            l[hour] = (data[list(data.keys())[0]]['s_d0'])

            #print(hour, data[list(data.keys())[0]]['s_d0'])
            
    return l
```

```python
# -*- coding: utf-8 -*-
from bs4 import BeautifulSoup
import csv
import json
import io
import requests

airbox_id = '28C2DDDD400A'
# airbox_id = '28C2DDDD4378'
school_name = '敦化國小'
# school_name = '健康國小'

file_name = './敦化國小PM25.csv'
# file_name = './健康國小PM25.csv'

date_month = '09'
date_day = '07'
data_written = []

while date_month != '01':
    res = requests.get('https://pm25.lass-net.org/data/history-date.php?device_id=' + airbox_id + '&date=2017-' + date_month + '-' + date_day)
    #res = requests.get('https://pm25.lass-net.org/data/history-date.php?device_id=28C2DDDD4378&date=2017-09-09')
    print('DB_0')
    soup = BeautifulSoup(res.text, "lxml")
    data_json = soup.select('body')[0].text
    data_json_parsed = json.loads(data_json)
    print(date_month + '/' + date_day)
    
    data_list = ['2017/' + date_month + '/' + date_day, school_name]
    data_list += clean(data_json_parsed)
    data_written.append(data_list)
    
    print(data_list)
    
    l = day_pass(date_month, date_day)
    date_month = l[0]
    date_day = l[1]
    
    

with io.open(file_name, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(data_written)
    f.close()
```
